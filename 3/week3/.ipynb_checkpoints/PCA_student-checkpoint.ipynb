# TODO 2

PCA  .

## 
1. normalize   
2.  numpy    
    - ```numpy.mean()```
    - ```numpy.linalg.eig()```
    - ```numpy.dot()```

````python
def student_pca(X, n_components=2):
    pca_results = None
    # TODO -->
    Xmean = X - np.mean(X, axis=0)
     
    matCov = np.cov(Xmean, rowvar=False)
     
    eigenVals, eigenVecs = np.linalg.eigh(matCov)
     
    sortIndex = np.argsort(eigenVals)[::-1]
    sortEigenVal = eigenVals[sortIndex]
    sortEigenVecs = eigenVecs[:,sortIndex]
     
    subsetEigenVec = sortEigenVecs[:,0:n_components]
     
    pca_results = np.dot(subsetEigenVec.transpose(), Xmean.transpose()).transpose()
    # <--
    return pca_results

pca_student_score = student_pca(X, n_components=2)
````

````python
show2D(pca_student_score, Y, target_names)
````

````python
def student_pca2(X, n_components=2):
    pca_results = None
    # TODO -->
    Xmean = X - np.mean(X, axis=0)
     
    matCov = np.cov(Xmean, rowvar=False)
     
    eigenVals, eigenVecs = np.linalg.eigh(matCov)
     
    sortIndex = np.argsort(eigenVals)[::-1]
    sortEigenVal = eigenVals[sortIndex]
    sortEigenVecs = eigenVecs[:,sortIndex]
     
    subsetEigenVec = sortEigenVecs[:,0:n_components]
     
    pca_results = np.dot(subsetEigenVec.transpose(), Xmean.transpose()).transpose()
    # <--
    return pca_results, subsetEigenVec

def student_mean_error(x, recon):
    return np.mean(pow(x - recon, 2))
````

````python
# TODO --> 
n_arr = [2, 3, 4, 32]

for n in n_arr:
    X = digits.data
    Xmean = np.mean(X, axis=0)
    
    pca, eigenVecs = student_pca2(X, n_components = n)    
    recon = np.dot(pca, eigenVecs.transpose()) + Xmean
    rmse = student_mean_error(X, recon)
    print(n, ": MSE Error:", rmse)
    
    # image
    n_samples = recon.shape[0]
    images = recon.reshape((n_samples, -1))
    _, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
    for ax, image in zip(axes, images):
        ax.set_axis_off()
        image = image.reshape(8, 8)
        ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
````
